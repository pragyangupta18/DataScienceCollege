Provide summary statistics (mean, median, minimum, maximum, standard deviation) for a dataset (age, income etc.) with numeric variables grouped by one of the qualitative (categorical) variable. For example, if your categorical variable is age groups and quantitative variable is income, then provide summary statistics of income grouped by the age groups. Create a list that contains a numeric value for each response to the categorical variable.

```python
from google.colab import files
import pandas as pd

uploaded = files.upload()
for filename in uploaded.keys():
    dataset_path = filename
df = pd.read_csv(dataset_path)
df
```

```python
import pandas as pd
import random

def generate_Adult(age):
    if age >= 27:
        return 'Old'
    else:
        return 'Young'

print(df.columns)

df['Age Range'] = df['Age '].apply(generate_Adult)

df
```

```python
# print(df.columns)
# income_mean = df['Income '].mean()
# print(f"Mean Income: {income_mean}")
```

```python
mean_salary = df.groupby('Age Range')['Income '].mean()
print(mean_salary)
```

```python
median_salary = df.groupby('Age Range')['Income '].median()
print(median_salary)
```

```python
df['Age Range'] = df['Age '].apply(generate_Adult)
mode_salary = df.groupby('Age Range')['Income '].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
print(mode_salary)
```

```python
Std_salary = df.groupby('Age Range')['Income '].std()
print(Std_salary)
```

```python
Count = df.groupby('Age Range')['Income '].count()
print(Count)
```

```python
variance_salary = df.groupby('Age Range')['Income '].var()
print(variance_salary)
```

```python
range_salary = df.groupby('Age Range')['Income '].apply(lambda x: x.max() - x.min())
print(range_salary)
```

```python
mid_range_salary = df.groupby('Age Range')['Income '].apply(lambda x: (x.max() + x.min()) / 2)
print(mid_range_salary)
```
