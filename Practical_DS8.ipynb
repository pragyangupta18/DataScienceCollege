Dist

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['age'].dropna(), kde=True, color='blue')
plt.title('Distribution Plot for Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Joint

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.jointplot(x='age', y='fare', data=titanic_data, kind='scatter', color='green')
plt.suptitle('Joint Plot of Age vs Fare', y=1.02)
plt.show()
```

Rug

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.rugplot(titanic_data['age'].dropna(), height=0.5, color='purple')
plt.title('Rug Plot for Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

Bar

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')


sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))
sns.barplot(x='pclass', y='fare', data=titanic_data, palette='muted')
plt.title('Bar Plot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.show()
```

Count

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.countplot(x='pclass', data=titanic_data, palette='pastel')
plt.title('Count Plot of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

Box

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(x='pclass', y='age', data=titanic_data, palette='dark')
plt.title('Box Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

Violen

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic_data = sns.load_dataset('titanic')


sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.violinplot(x='pclass', y='age', data=titanic_data, palette='deep')
plt.title('Violin Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

Strip

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')

sns.set(style="whitegrid")

# Strip Plot for the distribution of ages in each class
plt.figure(figsize=(10, 6))
sns.stripplot(x='pclass', y='age', data=titanic_data, jitter=True, palette='pastel', size=6)
plt.title('Strip Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

Swarm

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.stripplot(x='pclass', y='age', data=titanic_data, jitter=True, palette='pastel', size=6)
plt.title('Strip Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

Heat

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')

correlation_matrix = titanic_data.corr()

sns.set(style="whitegrid")

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heat Map of Correlation Matrix')
plt.show()
```

Cluster

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')

subset_data = titanic_data[['survived', 'pclass', 'age', 'fare']]

correlation_matrix = subset_data.corr()

sns.set(style="whitegrid")

plt.figure(figsize=(10, 8))
sns.clustermap(correlation_matrix, annot=True, cmap='viridis', linewidths=.5)
plt.title('Cluster Map of Correlation Matrix')
plt.show()
```

# Write a code to check how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a histogram.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['fare'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```
