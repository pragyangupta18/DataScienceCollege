# Data Formatting and Data Normalization

Data formatting

```python
import pandas as pd

# Create an unformatted dataset
data = {'Name': ['John Doe', 'Jane Smith', 'Bob Johnson'],
        'Age': [25, 30, 22],
        'Salary': [50000, 60000, 45000]}

df = pd.DataFrame(data)

# Display the unformatted dataset
print("Unformatted Dataset:")
print(df)
print("\n")

# Format the data
df['Name'] = df['Name'].apply(lambda x: x.upper())  # uppercase
df['Salary'] = df['Salary'].apply(lambda x: '${:,.2f}'.format(x))  # Add currency symbol and format salary
df['Formatted Age'] = df['Age'].apply(lambda x: f'{x} years old')  # Add a new column for formatted age
df = df.drop('Age', axis=1)

# Display the formatted dataset
print("Formatted Dataset:")
print(df)
```

Data Normalization

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

#data
data = {'Name': ['John Doe', 'Jane Smith', 'Bob Johnson'],
        'Age': [25, 30, 22],
        'Salary': [50000, 60000, 45000]}

data dictionary to a DataFrame
df = pd.DataFrame(data)

# Extract the numerical columns for normalization
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
X = df[numeric_columns]

# Display the original data
print("Before normalization:")
print(df)

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the data using the scaler
X_normalized = scaler.fit_transform(X)

# Replace the original numerical columns with the normalized values
df[numeric_columns] = X_normalized

# Display the data after normalization
print("\nAfter normalization:")
print(df)
```
