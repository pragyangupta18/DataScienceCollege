# Create an “Academic performance” dataset of students and perform the following operations using Python (Conditions Given in Problem Statement). Scan all variables for missing values and inconsistencies. If there are missing values and/or inconsistencies, use any of the suitable techniques to deal with them.


```python
import numpy as np
import pandas as pd
import random
from datetime import datetime, timedelta

random.seed(42)

num_instances = 31

data = {
    'Math_Score': [random.randint(60, 80) for _ in range(num_instances)],
    'Reading_Score': [random.randint(75, 95) for _ in range(num_instances)],
    'Writing_Score': [random.randint(60, 80) for _ in range(num_instances)],
    'Club_Join_Date': [(datetime(2018, 1, 1) + timedelta(days=random.randint(0, 1095))).strftime('%Y-%m-%d') for _ in range(num_instances)],
    'Placement_Score': [random.randint(75, 100) for _ in range(num_instances)],
}

df = pd.DataFrame(data)


def generate_offer_count(score):
    if score >= 90:
        return random.randint(4, 5)
    elif score >= 70:
        return random.randint(2, 3)
    else:
        return random.randint(0, 1)

df['Placement_Offer_Count'] = df['Placement_Score'].apply(generate_offer_count)

df.to_csv('Students_Performance_Placement.csv', index=False)

print("Dataset created and saved successfully.")

df
```

```python
df_copy = df.copy()

null_percentage = 0.3

for column in ['Math_Score', 'Writing_Score', 'Placement_Score']:
    num_nulls = int(num_instances * null_percentage)
    null_indices = random.sample(range(num_instances), num_nulls)
    df_copy.loc[null_indices, column] = np.nan

df_copy
```

```python
df_copy.isnull()
```

```python
df_copy.dropna(subset=['Writing_Score'], inplace=True)
df_copy
```

```python
df_copy.fillna(df_copy.mean(), inplace=True)
df_copy
```

```python
import pandas as pd

df_outlier = df.copy()

df_outlier['Math_Score'] = pd.to_numeric(df_outlier['Math_Score'], errors='coerce')

math_score_threshold = 78

math_score_outliers = df_outlier['Math_Score'] > math_score_threshold

mean_math_score = df_outlier.loc[~math_score_outliers, 'Math_Score'].mean()

df_outlier.loc[math_score_outliers, 'Math_Score'] = mean_math_score

print("Updated DataFrame with replaced outliers:")
print(df_outlier)
```

```python
import pandas as pd
df_outlier_mode = df.copy()

df_outlier_mode['Math_Score'] = pd.to_numeric(df_outlier_mode['Math_Score'], errors='coerce')

math_score_threshold = 78
math_score_mode = df_outlier_mode['Math_Score'].mode()[0]

mask = df_outlier_mode['Math_Score'] > math_score_threshold

df_outlier_mode.loc[mask, 'Math_Score'] = math_score_mode

print("Updated DataFrame :")
print(df_outlier_mode)
```

```python
import pandas as pd

df_outlier = df.copy()

df_outlier['Math_Score'] = pd.to_numeric(df_outlier['Math_Score'], errors='coerce')

math_score_threshold = 78
math_score_median = df_outlier['Math_Score'].median()

mask = df_outlier['Math_Score'] > math_score_threshold

df_outlier_median = df_outlier.copy()

df_outlier_median.loc[mask, 'Math_Score'] = math_score_median

print("Updated DataFrame with replaced outliers and median:")
print(df_outlier_median)
```

```python
import pandas as pd
import numpy as np

# Assuming your original dataset is named df
data = {'Age': [22, 25, 30, 35, 40],
        'Income': [50000, 60000, 75000, 80000, 90000],
        'Score': [85, 88, 92, 78, 95]}

df = pd.DataFrame(data)
# Smoothing
df['Smoothed_Score'] = df['Score'].rolling(window=2).mean()

# Aggregation
aggregated_df = df.groupby('Age').agg({'Score': 'sum'}).reset_index()

# Generalization
bins = [0, 30, 100]
labels = ['Young', 'Old']
df['Age_Category'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Normalization - Min-max normalization
df['Normalized_Income_MinMax'] = (df['Income'] - df['Income'].min()) / (df['Income'].max() - df['Income'].min())

# Normalization - Z-score normalization
df['Normalized_Score_ZScore'] = (df['Score'] - df['Score'].mean()) / df['Score'].std()

# Normalization - Decimal scaling
df['Normalized_Age_DecimalScaling'] = df['Age'] / 10

# Displaying the modified DataFrame
print("Original DataFrame:")
print(df)

print("\nAggregated DataFrame:")
print(aggregated_df)
```
